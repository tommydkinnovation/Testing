# Examples of inserting an image

## Insert an image from a Folder

[![title](img/gmit_logo.png)
](https://student.gmit.ie/)

```python
import urllib.request
import io
import PIL.Image
URL = 'http://www.gmit.ie/sites/all/themes/gmitpublic/images/gmit_logo.png'


with urllib.request.urlopen(URL) as url:
     imagefile = io.BytesIO(url.read())
     image = PIL.Image.open(imagefile)
     imagebw = image.convert('L')

display(image)
```

## Making the image Back and White

```python
import urllib.request
import io
import PIL.Image
URL = 'http://www.gmit.ie/sites/all/themes/gmitpublic/images/gmit_logo.png'

with urllib.request.urlopen(URL) as url:
     imagefile = io.BytesIO(url.read())
     image = PIL.Image.open(imagefile)
     imagebw = image.convert('L')

display(imagebw)
```
